# ***Problem Statement 1: Employee Performance Bonus Eligibility***


```python
# employee bonus eligibility thing -- need to find who got max score
# idea: first grab highest score, then check who all have it

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

def find_top_performers(emp_dict):
    # ok first step -> find max score from all values
    highest_score = max(emp_dict.values())  # this gives biggest score in dict

    # now collect employees who got this score
    top_employees = []  # empty list for storing names
    for name, score in emp_dict.items():
        # if score matches highest, add name
        if score == highest_score:
            top_employees.append(name)

    return top_employees, highest_score


# calling function
top_names, score = find_top_performers(employees)

# join names nicely for printing
names_string = ", ".join(top_names)

print(f"Top Performers Eligible for Bonus: {names_string} (Score: {score})")
```

# ***Problem Statement 2: Search Query Keyword Analysis***


```python
# search keyword analysis -- need to count repeated words from a sentence
# steps in head: lowercase -> remove punctuation -> split -> count -> filter

import string  # using this to strip punctuation fast

query = "Buy mobile phone buy phone online"

def analyze_keywords(sentence):
    # convert everything to lowercase first (case shouldn't matter)
    sentence = sentence.lower()

    # remove punctuation (just in case user typed weird stuff like !,. etc)
    # translating punctuation to nothing basically deletes it
    cleaned = sentence.translate(str.maketrans('', '', string.punctuation))

    # split sentence into words
    words = cleaned.split()  # default split on spaces

    # count frequency of each word
    freq = {}
    for word in words:
        # if word already seen -> increase count
        # else start count at 1
        freq[word] = freq.get(word, 0) + 1

    # now keep only words that appear more than once
    repeated = {}
    for word, count in freq.items():
        if count > 1:
            repeated[word] = count

    return repeated


# calling function
result = analyze_keywords(query)

print(result)
```

# ***Problem Statement 3: Sensor Data Validation***


```python
# sensor data validation thing -- need to pick only even readings
# plan: loop with index, check even, store (index, value)

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

def get_valid_readings(readings):
    valid = []  # list to store good readings

    # enumerate gives both index (hour) and value
    for hour, value in enumerate(readings):
        # even check -> if remainder 0 when divided by 2
        if value % 2 == 0:
            # store tuple of (hour, value)
            valid.append((hour, value))

    return valid


# calling function
valid_readings = get_valid_readings(sensor_readings)

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# ***Problem Statement 4: Email Domain Usage Analysis***

```python
# email domain analysis -- count domains + find percentage usage
# rough plan: extract domain -> count -> convert to %

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

def analyze_domains(email_list):
    domain_count = {}  # stores how many per domain

    # loop through each email and grab domain part
    for email in email_list:
        # split at '@' and take right side
        domain = email.split("@")[1]

        # increase count (or start at 1 if new)
        domain_count[domain] = domain_count.get(domain, 0) + 1

    total_users = len(email_list)  # total emails for percentage calc

    # now convert counts to percentages
    domain_percentage = {}
    for domain, count in domain_count.items():
        # percentage formula -> (part/total)*100
        percent = (count / total_users) * 100
        domain_percentage[domain] = percent

    return domain_percentage


# calling function
result = analyze_domains(emails)

# printing nicely
for domain, percent in result.items():
    print(f"{domain}: {int(percent)}%")
```

# ***Problem Statement 5: Sales Spike Detection***


```python
# sales spike detection -- need average first, then find days way above it
# logic in head: avg -> threshold (avg + 30%) -> compare each day

sales = [1200, 1500, 900, 2200, 1400, 3000]

def detect_sales_spikes(sales_data):
    # calculate average sales
    total = sum(sales_data)  # add all sales
    count = len(sales_data)  # number of days
    average = total / count  # basic avg formula

    # figure out spike threshold (30% above avg)
    threshold = average * 1.30  # same as avg + 30%

    spikes = []  # store spike days here

    # loop with day numbers (starting from 1, not 0)
    for day, value in enumerate(sales_data, start=1):
        # if value crosses threshold -> it's a spike
        if value > threshold:
            spikes.append((day, value))

    return spikes


# calling function
spike_days = detect_sales_spikes(sales)

# printing result
for day, value in spike_days:
    print(f"Day {day}: {value}")
```

# ***Problem Statement 6: Duplicate User ID Detection***


```python
# duplicate user id detection -- basically count repeats and show only duplicates
# plan: count all ids first, then filter ones with count > 1

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

def find_duplicates(ids):
    counts = {}  # dict to store frequency of each id

    # counting how many times each id shows up
    for uid in ids:
        # if already exists bump count, else start at 1
        counts[uid] = counts.get(uid, 0) + 1

    duplicates = {}  # store only repeated ids

    # filter out ids that appear more than once
    for uid, freq in counts.items():
        if freq > 1:
            duplicates[uid] = freq

    return duplicates


# calling function
dupes = find_duplicates(user_ids)

# printing result in required format
for uid, freq in dupes.items():
    print(f"{uid} → {freq} times")
```
