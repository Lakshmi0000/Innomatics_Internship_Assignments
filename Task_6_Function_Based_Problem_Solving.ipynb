# ***Problem Statement 1:Smart Parking Lot Management System***


```python
# smart parking lot management system
# idea:
# - loop through logs
# - increase count for "IN"
# - decrease count for "OUT"
# - track peak usage also
# - check if exceeds capacity

def manage_parking(capacity, vehicle_logs):
    current_parked = 0     # vehicles currently inside
    peak_usage = 0         # highest number reached anytime

    # processing each log entry one by one
    for log in vehicle_logs:

        if log == "IN":
            current_parked += 1   # car entered

        elif log == "OUT":
            # making sure count doesn't go negative (just safety)
            if current_parked > 0:
                current_parked -= 1

        # update peak usage if current exceeds previous peak
        if current_parked > peak_usage:
            peak_usage = current_parked

    # check capacity condition
    if current_parked > capacity:
        status = "Capacity Exceeded - Alert!"
    else:
        status = "Available"

    return current_parked, peak_usage, status


# sample input
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# calling function
current, peak, parking_status = manage_parking(parking_capacity, vehicle_logs)

# printing required output
print(f"Currently Parked Vehicles: {current}")
print(f"Parking Status: {parking_status}")
```

# ***Problem Statement 2:Online Food Delivery Time Estimator***


```python
# online food delivery time estimator
# logic in mind:
# base time depends on distance
# then add extra minutes depending on traffic + weather
# final ETA = base + delays

def estimate_delivery_time(distance_km, traffic_level, weather_condition):

    # assuming average speed ~ 30 km/h (rough city assumption)
    # so base time = distance / speed * 60 (convert to minutes)
    base_time = (distance_km / 30) * 60

    total_time = base_time  # start with base

    # traffic impact
    if traffic_level == "High":
        total_time += 20   # heavy delay
    elif traffic_level == "Medium":
        total_time += 10
    # low traffic no extra delay

    # weather impact
    if weather_condition == "Rainy":
        total_time += 15
    elif weather_condition == "Stormy":
        total_time += 25
    # clear weather no extra delay

    return round(total_time)


# sample input
distance = 8
traffic = "High"
weather = "Rainy"

# calling function
eta = estimate_delivery_time(distance, traffic, weather)

# printing required output
print(f"Estimated Delivery Time: {eta} minutes")
```

# ***Problem Statement 3:Movie Theatre Seat Occupancy***

```python
# movie theatre seat occupancy analyzer
# plan:
# 1) count booked seats
# 2) calculate occupancy %
# 3) decide show status
# 4) suggest extra show if needed

def analyze_seat_occupancy(total_seats, booked_seats_list):

    booked_count = 0  # will count how many seats booked

    # counting booked seats manually (using loop as hint says)
    for seat in booked_seats_list:
        if seat == 1:   # assuming 1 means booked
            booked_count += 1

    # calculate occupancy percentage
    occupancy_percentage = (booked_count / total_seats) * 100

    # deciding show status
    if occupancy_percentage == 100:
        status = "Housefull"
        suggestion = "Consider Opening Additional Show"

    elif occupancy_percentage >= 75:
        status = "Almost Full"
        suggestion = "Monitor Bookings"

    else:
        status = "Seats Available"
        suggestion = "No Additional Show Needed"

    return round(occupancy_percentage), status, suggestion


# sample input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# calling function
occupancy, show_status, extra_suggestion = analyze_seat_occupancy(total_seats, booked_seats)

# printing required output format
print(f"Occupancy: {occupancy}%")
print(f"Show Status: {show_status}")
```

# ***Problem Statement 4:Cloud Server Load Classification System***

```python
# cloud server load classification system
# steps in head:
# 1) calculate average CPU using loop
# 2) classify based on given ranges

def classify_server_load(cpu_readings):

    total_cpu = 0  # to accumulate CPU values

    # calculating total manually (using loop as hinted)
    for reading in cpu_readings:
        total_cpu += reading   # adding each reading

    # calculate average
    average_cpu = total_cpu / len(cpu_readings)

    # classification logic
    if average_cpu < 50:
        status = "Normal"

    elif average_cpu <= 80:
        # this covers 50% to 80%
        status = "Warning"

    else:
        # above 80%
        status = "Critical"

    return round(average_cpu), status


# sample input
cpu_data = [45, 60, 70, 85, 90]

# calling function
avg_cpu, server_status = classify_server_load(cpu_data)

# printing required output
print(f"Average CPU Load: {avg_cpu}%")
print(f"Server Status: {server_status}")
```

# ***Problem Statement 5:Smart Classroom Resource Usage Monitor***

```python
# smart classroom resource usage monitor
# plan in head:
# - loop through dictionary
# - check which resource crossed safe usage limit
# - collect overused ones
# assuming overuse threshold = 8 hours (daily limit kind of thing)

def monitor_resource_usage(resource_dict):

    overused = []   # will store names of overused resources
    threshold = 8   # fixed safe usage limit (hours)

    # checking each resource
    for resource, hours in resource_dict.items():

        # if usage crosses threshold -> mark as overused
        if hours > threshold:
            overused.append(resource)

    # energy alert if any resource overused
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    return overused, alert


# sample input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# calling function
overused_resources, energy_alert = monitor_resource_usage(resource_usage)

# printing required output format
if overused_resources:
    print(f"Overused Resources: {', '.join(overused_resources)}")
else:
    print("Overused Resources: None")

print(f"Energy Alert: {energy_alert}")
```

# ***Problem Statement 6: Online Event Registration Capacity Controller***

```python
# online event registration capacity controller
# plan:
# - simulate registrations using loop
# - allow only upto capacity
# - extra go to waitlist
# - decide registration status

def manage_event_registrations(capacity, total_registrations):

    confirmed = 0   # confirmed seats
    waitlisted = 0  # extra users

    # simulate each registration attempt
    for i in range(total_registrations):

        if confirmed < capacity:
            confirmed += 1   # seat available
        else:
            waitlisted += 1  # capacity full -> waitlist

    # registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status


# sample input
event_capacity = 100
registrations = 105

# calling function
confirmed_count, waitlist_count, registration_status = manage_event_registrations(
    event_capacity, registrations
)

# printing required output
print(f"Confirmed Registrations: {confirmed_count}")
print(f"Waitlisted Users: {waitlist_count}")
print(f"Registration Status: {registration_status}")
```
