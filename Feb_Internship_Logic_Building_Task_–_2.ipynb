# ***Problem 1: Unique Words in a Sentence***

```python
# Ask the user to enter a sentence
sentence = input("Enter sentence: ")

# Split the sentence into individual words using spaces as separators
words = sentence.split()

# Convert the list of words into a set to remove duplicates
# A set automatically keeps only unique elements
unique_words = set(words)

# Count the number of unique words
unique_count = len(unique_words)

# Display the results
print("Unique words count:", unique_count)
print("Unique words:", unique_words)
```

# ***Problem 2: Highest Salary from Employee Data***

```python
# Create an empty dictionary to store employee data
employee_data = {}

# Ask how many employees we want to enter
n = int(input("Enter number of employees : "))

# Loop n times to collect employee details
for i in range(n):
    # Take employee name and salary as input
    name = input("Enter employee name : ")
    salary = int(input("Enter employee salary : "))

    # Store the name as key and salary as value in the dictionary
    employee_data[name] = salary

# Find the maximum salary from all stored salaries
max_salary = max(employee_data.values())

# Loop through the dictionary to find who has the highest salary
for name, salary in employee_data.items():
    # Compare each salary with the maximum salary
    if salary == max_salary:
        print(f"Highest Salary : {name} - {salary}")
```

# ***Problem 3: Find Maximum and Minimum Values***

```python
# taking input from user (space separated numbers)
numbers = list(map(int, input("Enter numbers: ").split()))

# setting first element as max and min for now
# we’ll compare others with this
max_value = numbers[0]
min_value = numbers[0]

# going through each number in the list one by one
for num in numbers:

    # if we find something bigger than current max, update it
    if num > max_value:
        max_value = num

    # same idea for minimum — if smaller number found, update
    if num < min_value:
        min_value = num

# printing final results
print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# ***Problem 4: Count Products Above a Price Threshold***

```python
# taking product prices
prices = list(map(int, input("Enter product prices: ").split()))

# counter to keep track of how many are above 1000
count = 0

# checking each price one by one
for price in prices:

    # if price is greater than 1000, increase the count
    if price > 1000:
        count += 1

# final result
print("Products above 1000:", count)
```

# ***Problem 5: Calculate Attendance Percentage***

```python
# taking attendance input (P and A separated by spaces)
attendance = input("Enter attendance: ").upper().split()

# total number of days recorded
total_days = len(attendance)

present = 0  # counter to track how many days are present

# going through each attendance entry
for mark in attendance:

    # if it's P, count it as present
    if mark == "P":
        present += 1

# calculating attendance percentage
attendance_percentage = (present / total_days) * 100

# printing the result
print(f"Attendance Percentage: {attendance_percentage}")
```

# ***Problem 7: Count Character Frequency***

```python
# taking characters as input and converting into list
original_list = list(input("Enter the characters: "))

data = {}  # dictionary to store frequency

# getting only unique characters to avoid recounting
unique_char = set(original_list)

# looping through each unique character
for unique_numbers in unique_char:
    count = 0  # reset counter for each character

    # checking how many times it appears in original list
    for char in original_list:
        if char == unique_numbers:
            count += 1

    # storing final count in dictionary
    data[unique_numbers] = count

# printing result
print(data)
```

# ***Problem 9: Check if a Key Exists in a Dictionary***

```python
import ast  # for safely taking dictionary input

# taking dictionary input from user
employee_data = ast.literal_eval(input("Enter dictionary: "))

# asking which key we want to check
key = input("Enter key to search: ")

# checking if the key exists in dictionary
if key in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# ***Problem 10: Calculate Average Marks***

```python
# taking marks as space-separated input
marks = list(map(int, input("Enter marks: ").split()))

# finding total marks by adding everything
total = sum(marks)

# number of subjects/marks entered
count = len(marks)

# calculating average
average = total / count

# printing final result
print("Average Marks:", average)
```
