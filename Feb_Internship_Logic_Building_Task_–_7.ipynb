# ***Problem 1: Smart Light Controller***
Problem Statement:
Create a class to control a smart light that can be turned ON or OFF and display its status.
🌍 Real-Time Use
Smart home automation systems
📥 Sample Input
Light Name: Bedroom Light
Action: ON
📤 Expected Output
Bedroom Light is ON


```python
# class for controlling a smart light
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # default state... light usually starts off

    def turn_on(self):
        # turning light ON
        self.status = "ON"

    def turn_off(self):
        # turning light OFF... simple
        self.status = "OFF"

    def display_status(self):
        # printing current state of the light
        print(f"{self.name} is {self.status}")


# --- main program part ---

# taking input from user
light_name = input("Light Name: ")
action = input("Action: ")

# creating smart light object
light = SmartLight(light_name)   # ok object created

# checking what action user wants
if action.upper() == "ON":
    light.turn_on()      # call ON method
elif action.upper() == "OFF":
    light.turn_off()     # call OFF method
else:
    print("Invalid action")   # just in case user types something weird

# finally show light status
light.display_status()
```

# ***Problem 2: Employee ID Card System***
 Problem Statement:
Create a class that stores employee name, ID, and department and displays ID card details.
Real-Time Use
Corporate HR systems
📥 Sample Input
Employee Name: Rahul
Employee ID: EMP102
Department: IT

📤 Expected Output
Employee ID Card
Name: Rahul
ID: EMP102
Department: IT


```python
# class representing an Employee ID Card
class EmployeeIDCard:

    def __init__(self, name, emp_id, department):
        # storing employee details here
        self.name = name
        self.emp_id = emp_id
        self.department = department   # dept like IT, HR etc

    def display_card(self):
        # print the ID card info... simple formatting
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# ---- main program part ----

# taking inputs from user
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# create object of EmployeeIDCard
employee = EmployeeIDCard(name, emp_id, department)   # object created... holds all info

# display the card details
employee.display_card()   # calling method to print details
```

# ***Problem 3: Mobile Contact Record***
Problem Statement:
Create a class to store contact name and phone number and display contact information.
Real-Time Use
Mobile contact applications
📥 Sample Input
Contact Name: Anita
Phone Number: 9876543210
📤 Expected Output
Contact Saved
Name: Anita
Phone: 9876543210


```python
# class for storing a mobile contact
class Contact:

    def __init__(self, name, phone):
        # saving contact info here
        self.name = name
        self.phone = phone   # phone number stored as given

    def display_contact(self):
        # show saved contact details
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# ----- main part of program -----

# taking input from user
contact_name = input("Contact Name: ")
phone_number = input("Phone Number: ")

# create contact object
contact1 = Contact(contact_name, phone_number)   # ok object created with data

# display the saved contact
contact1.display_contact()   # calling method to print info
```

# ***Problem 4: Product Price Tag Generator***
Problem Statement:
Create a class that stores product name and price and prints a formatted price tag.
 Real-Time Use
Retail billing systems
📥 Sample Input
Product Name: Headphones
Price: 2499
📤 Expected Output
Product: Headphones
Price: ₹2499


```python
# class for generating a product price tag
class Product:

    def __init__(self, name, price):
        # storing product details here
        self.name = name
        self.price = price   # price as entered... assuming rupees

    def print_price_tag(self):
        # printing the formatted price tag
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")   # rupee symbol added while printing


# ---- main program starts here ----

# taking input from user
product_name = input("Product Name: ")
price = input("Price: ")

# create product object
product1 = Product(product_name, price)   # ok product object ready

# now print the price tag
product1.print_price_tag()   # calling method to show output
```

# ***Problem 6: Delivery Address Manager***
Problem Statement:
Create a class that stores customer name and delivery address and prints delivery details.
Real-Time Use
Courier and logistics platforms
📥 Sample Input
Customer Name: Suman
Address: Hyderabad
📤 Expected Output
Delivery Details
Customer: Suman
Address: Hyderabad


```python
# class to manage delivery address details
class Delivery:

    def __init__(self, customer_name, address):
        # storing customer info
        self.customer_name = customer_name
        self.address = address   # delivery address where package goes

    def show_delivery_details(self):
        # printing delivery details in required format
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")   # simple print... nothing fancy


# ---- main program ----

# taking inputs from user
customer_name = input("Customer Name: ")
address = input("Address: ")

# create delivery object
delivery1 = Delivery(customer_name, address)   # object created with customer data

# display delivery info
delivery1.show_delivery_details()   # calling method to print delivery details
```
