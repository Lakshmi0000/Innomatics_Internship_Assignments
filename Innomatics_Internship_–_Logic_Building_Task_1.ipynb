# 1. ***User Login Check***

```python
# Stored credentials (these represent the correct login details)
correct_username = "admin"
correct_password = "1234"

# Take login details dynamically from the user
username = input("Enter username: ").strip()
password = input("Enter password: ").strip()

# Check if the entered credentials match the stored ones
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. ***Pass / Fail Analyzer***

```python
def analyze_results(marks):
    # Counters to keep track of pass and fail students
    pass_count = 0
    fail_count = 0

    for score in marks:
        # A student passes if marks are 50 or above
        if score >= 50:
            pass_count += 1
        else:
            fail_count += 1

    # Display the final results in a clear format
    print("Result Analysis")
    print("----------------")
    print(f"Total Students : {len(marks)}")
    print(f"Passed         : {pass_count}")
    print(f"Failed         : {fail_count}")


# Given marks list
marks = [45, 78, 90, 33, 60]

analyze_results(marks)
```

# 3. ***Simple Data Cleaner***

```python
def clean_names(names):
    cleaned_names = []

    for name in names:
        # Remove extra spaces and convert to lowercase
        cleaned_name = name.strip().lower()

        # Add the cleaned name to the new list
        cleaned_names.append(cleaned_name)

    # Print the final cleaned list
    print("Cleaned Names:", cleaned_names)


# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Run the data cleaning function
clean_names(names)
```

# 4. ***Message Length Analyzer***

```python
def analyze_messages(messages):

    for message in messages:
        # Calculate the length of the current message
        length = len(message)

        # Print the message and its length
        print(f"Message: '{message}',  Length: {length}")

        # Flag messages that are longer than 10 characters
        if length > 10:
            print(" This message is longer than 10 characters")

        print()


# Given list of messages
messages = ["Hi", "Welcome to the platform", "OK"]

#Run the message analyzer
analyze_messages(messages)
```

# ***5. Error Message Detector***

```python
def detect_errors(logs):
    # Initialize a counter to track how many ERROR messages appear
    error_count = 0

    for entry in logs:
        # Check if the current entry is an ERROR
        if entry == "ERROR":
            error_count += 1

    # Print the final error count clearly
    print("System Log Analysis")
    print("-------------------")
    print(f"Total ERROR entries: {error_count}")


#Given system logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


detect_errors(logs)
```
