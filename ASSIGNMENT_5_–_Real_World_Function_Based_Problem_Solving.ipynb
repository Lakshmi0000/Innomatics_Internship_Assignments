# ***Problem 1: Social Media – Post Engagement Analyzer***

```python
# social media post engagement analyzer
# goal -> calculate total likes and decide if post is viral or not
# rule stuck in head: >= 1000 means Viral, else Normal

def analyze_post_engagement(likes_list):
    # first calculate total likes (just summing manually using loop)
    total_likes = 0

    for likes in likes_list:
        # adding each like count one by one
        total_likes += likes   # same as total_likes = total_likes + likes

    # now decide post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# example data (can change this to test)
post_likes = [250, 300, 150, 200, 150]  # total = 1050

# calling function
total, post_status = analyze_post_engagement(post_likes)

# printing in required format
print(f"Total Likes: {total}")
print(f"Post Status: {post_status}")
```

# ***Problem 3: Agriculture – Rainfall Adequacy Checker***

```python
# rainfall adequacy checker -- need avg rainfall and compare with required level
# logic: sum using loop -> divide by count -> check condition

def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0  # will store total rainfall

    # calculate total using loop (not directly using sum, doing manually)
    for rain in rainfall_data:
        total_rainfall += rain   # adding one by one

    # calculate average
    average_rainfall = total_rainfall / len(rainfall_data)

    # check if rainfall is adequate or not
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average_rainfall, status


# example seasonal rainfall data
season_rainfall = [70, 75, 80, 65, 70]   # avg = 72
required_rainfall_level = 70

# calling function
avg, rainfall_status = check_rainfall(season_rainfall, required_rainfall_level)

# printing in required format
print(f"Average Rainfall: {int(avg)}")
print(f"Rainfall Status: {rainfall_status}")
```

# ***Problem 4: Social Media – Duplicate Account Detection***

```python
# duplicate account detection -- check if any username repeats
# hint said use set, so yeah comparing length of list vs set

def detect_duplicate_accounts(usernames):
    # convert list to set (set automatically removes duplicates)
    unique_users = set(usernames)

    # if lengths differ -> duplicates were there
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    return result


# example data (change to test)
accounts = ["ravi123", "anita98", "ravi123", "kiran77"]

# calling function
duplicate_status = detect_duplicate_accounts(accounts)

# printing in required format
print(f"Duplicate Accounts Found: {duplicate_status}")
```

# ***Problem 6: Agriculture – Premium Crop Price Filter***

```python
# premium crop price filter -- need to pick only prices above threshold
# rule: > 2000 means premium, else ignore (regular crop no need to store)

def filter_premium_crops(crop_prices, threshold):
    premium_list = []  # will store only premium crop prices

    # looping through each price manually (not using fancy stuff)
    for price in crop_prices:
        # check if price crosses threshold
        if price > threshold:
            premium_list.append(price)   # add to list if premium

    return premium_list


# example crop prices (can modify and test)
prices = [1500, 2500, 1800, 3200, 2000]
premium_threshold = 2000

# calling function
premium_crops = filter_premium_crops(prices, premium_threshold)

# printing in required format
print(f"Premium Crops: {premium_crops}")
```

# ***Problem 7: System Monitoring – Application Health Checker***

```python
# application health checker -- decide health based on error count
# rule in head:
# 0 errors -> Healthy
# <=5 errors -> Minor Issues
# >5 errors -> Critical Issues

def check_application_health(error_count):
    # checking conditions one by one (order matters here)

    if error_count == 0:
        status = "Healthy"

    elif error_count <= 5:
        # this means 1 to 5 errors basically
        status = "Minor Issues"

    else:
        # anything more than 5 comes here
        status = "Critical Issues"

    return error_count, status


# example error count (can change and test)
current_errors = 7

# calling function
errors, system_status = check_application_health(current_errors)

# printing in required format
print(f"Error Count: {errors}")
print(f"System Status: {system_status}")
```

# ***Problem 8: Banking – Daily Transaction Limit Checker***

# ***Problem 9: E-Learning – Student Attendance Eligibility System***

```python
# student attendance eligibility system -- calculate % and check eligibility
# plan: count total classes, count attended ones, compute percentage

def check_attendance(attendance_list):
    total_classes = len(attendance_list)   # total sessions conducted
    attended_classes = 0   # counter for attended classes

    # assuming 1 means present, 0 means absent
    for status in attendance_list:
        if status == 1:
            attended_classes += 1   # increase if present

    # calculate percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    # check eligibility condition
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return attendance_percentage, eligibility


# example attendance record (1 = present, 0 = absent)
attendance_data = [1, 1, 0, 1, 1]   # 4/5 present -> 80%

# calling function
percentage, exam_status = check_attendance(attendance_data)

# printing in required format
print(f"Attendance Percentage: {percentage}")
print(f"Exam Eligibility: {exam_status}")
```

# ***Problem 10: Smart Electricity Bill Analyzer***

```python
# smart electricity bill analyzer -- slab based billing system
# plan in head:
# 1) calculate bill according to slabs
# 2) classify usage based on total bill amount

def calculate_electricity_bill(units):
    total_bill = 0  # will accumulate bill here

    # applying slab logic carefully (don't mess up slabs)

    if units <= 100:
        # all units charged at 3
        total_bill = units * 3

    elif units <= 200:
        # first 100 at 3
        # remaining at 5
        total_bill = (100 * 3) + ((units - 100) * 5)

    else:
        # first 100 at 3
        # next 100 at 5
        # remaining at 7
        total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # now classify usage based on bill amount

    if total_bill < 500:
        usage_status = "Low Usage"

    elif total_bill <= 1500:
        # between 500 and 1500
        usage_status = "Moderate Usage"

    else:
        usage_status = "High Usage"

    return total_bill, usage_status


# example usage (can change and test)
units_consumed = 250

# calling function
bill_amount, usage_level = calculate_electricity_bill(units_consumed)

# printing result clearly
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Classification: {usage_level}")
```
